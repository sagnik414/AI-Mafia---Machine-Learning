```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.available
plt.style.use("seaborn")
```

```python
dfx = pd.read_csv('Linear_X_Train.csv')
dfy = pd.read_csv('Linear_Y_Train.csv')

dfx = dfx.values
dfy = dfy.values

x = dfx.reshape((-1,))
y = dfy.reshape((-1,))

```

```python
x = (x-x.mean())/x.std()
y = y
plt.scatter(x,y,color='red')
plt.show()
```

```python
# Gradient descent algorithm
```

```python
def hypothesis(x,theta):
    return theta[0] + theta[1]*x
```

```python
def error(x,y,theta):
    error = 0
    
    for i in range(x.shape[0]):
        
        hx = hypothesis(x[i],theta)
        
        error += (hx-y[i])**2
        
    return error
```

```python
def gradient(x,y,theta):
    
    grad = np.zeros((2,))
    m = x.shape[0]

    for i in range(m):
        hx = hypothesis(x[i],theta)
        grad[0] +=  (hx-x[i])
        grad[1] += (hx-y[i])*x[i]
        
    return grad
```

```python
def gradientdescentalgo(x,y,learning_rate=0.00001):
    # random theta value
    
    theta = np.array([0.7,1.0])
    
    max_iteration = 100 # it means after 100 interation we will get the optimal result 
    
    iter = 0
    
    error_list = []
    theta_list = []
    
    
    while(iter<=max_iteration):
        
        grad = gradient(x,y,theta)
        err = error(x,y,theta)
        error_list.append(err)
        theta_list.append(theta)
        theta[0] -=learning_rate*grad[0]
        theta[1] -=learning_rate*grad[1]
        
        iter+=1
        
    return theta,error_list,theta_list
```

```python
final_theta, error_list,theta_list = gradientdescentalgo(x,y)
```
